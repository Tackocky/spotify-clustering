```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("genres_v2.csv")
df.head()
```

```python
df.info()
```

```python
df2 = df.iloc[:,0:13]
df2=df2.drop(columns="type")
df2["easy_id"]=df.index+1
df2.head()
```
